```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from lib.city import *
from lib.visualize import *
from lib.tspga import GAtsp 
```

```python
city = gen_city_df(15, 0, 400)
city.head()
```

```python
distance = cal_distance_matrix(city)

plt.figure(figsize=(8, 8), dpi=80)
mask = np.zeros_like(distance)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(distance, square=True, annot=True, mask=mask, cmap='viridis')
plt.show()
```

```python
population_size = 20
best_parent = 5
nextgen_parent = 10
mutation_prob = 0.5
tournament_size = 3

max_generation = 250

ga = GAtsp(population_size=population_size, 
           best_parent=best_parent, 
           nextgen_parent=nextgen_parent, 
           mutation_prob=mutation_prob, 
           tournament_size=tournament_size)
ga.fit(city, distance)
```

```python
ga.run_ga(max_generation)
```
